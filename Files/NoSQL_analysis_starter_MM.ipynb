### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
hygiene_query = {"scores.Hygiene": 20}
results = establishments.find(hygiene_query)

# Use count_documents to display the number of documents in the result
hygiene_documents = establishments.count_documents(hygiene_query)
print("Number of documents in the result:", hygiene_documents)

# Display the first document in the results using pprint
first_document = results[0]
pprint(first_document)
```

```python
# Convert the result to a Pandas DataFrame
hygeine_df = pd.DataFrame(list(results))

# Display the number of rows in the DataFrame

hygeine_df_rows = len(hygeine_df)
print("Number of rows in the DataFrame:", hygeine_df_rows)

# Display the first 10 rows of the DataFrame
hygeine_df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
rating_query = {"LocalAuthorityName": {"$regex": "London", "$options": "i"}, "RatingValue": {"$gte": 4}}
rating_results = establishments.find(rating_query)

# Use count_documents to display the number of documents in the result
rating_documents = establishments.count_documents(rating_query)
print("Number of documents in the result:", rating_documents)

# Display the first document in the results using pprint

first_rating_document = rating_results[0]
pprint(first_rating_document)
```

```python
# Convert the result to a Pandas DataFrame
rating_df = pd.DataFrame(list(rating_results))

# Display the number of rows in the DataFrame
rating_df_rows = len(rating_df)
print("Number of rows in the DataFrame:", rating_df_rows)

# Display the first 10 rows of the DataFrame
rating_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
#Find Penang Flavours Collection
penang = establishments.find_one({"BusinessName": "Penang Flavours"})

#Extract Latiude and Longitude
penang_latitude = penang['geocode']['latitude']
penang_longitude = penang['geocode']['longitude']

# Define the range around the target coordinate (0.1 degree)
range_degrees = 0.1

# Define the latitude and longitude range
latitude_range = (penang_latitude - range_degrees, penang_latitude + range_degrees)
longitude_range = (penang_longitude - range_degrees, penang_longitude + range_degrees)

# Construct the query to find top 5 establishments with RatingValue of 5, sorted by lowest hygiene score within the specified range
penang_query = {
    "RatingValue": 5,
    "geocode.latitude": {"$gte": latitude_range[0], "$lte": latitude_range[1]},
    "geocode.longitude": {"$gte": longitude_range[0], "$lte": longitude_range[1]},
    #"scores.Hygiene": 
}

# Find the top 5 establishments with RatingValue of 5, sorted by lowest hygiene score
pf_results = list(establishments.find(penang_query).sort({"scores.Hygiene": -1}).limit(5))
#list(establishments.find(penang_query).sort("scores.Hygiene", 1).limit(5))

pprint(pf_results)
```

```python
# Convert result to Pandas DataFrame
pf_df = pd.DataFrame(list(pf_results))
pf_df
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that:
# 1. Matches establishments with a hygiene score of 0
match_query = {
    '$match': {
        'scores.Hygiene': 0
    }
}

# 2. Groups the matches by Local Authority
group_query = {
    '$group': {
        '_id': '$LocalAuthorityName',
        'count': {'$sum': 1}
    }
}

# 3. Sorts the matches from highest to lowest
sort_query = {
    '$sort': {'count': -1}
}

# Print the number of documents in the result
pipeline = [match_query, group_query, sort_query]
result = list(establishments.aggregate(pipeline))
# Print the first 10 results
pprint(result[0:10])
```

```python
# Convert the result to a Pandas DataFrame
# Convert result to Pandas DataFrame
hygeine_zero_df = pd.DataFrame(list(result))

# Display the number of rows in the DataFrame
hygeine_zero_df

# Display the first 10 rows of the DataFrame
hygeine_zero_df.head(10)
```
